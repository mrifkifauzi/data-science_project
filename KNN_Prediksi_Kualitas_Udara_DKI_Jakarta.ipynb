```python
#Membuat design matrix X dan target vector y

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

#Membuat design matrix X dengan scaling data menggunakan StandardScaler
X = pd.DataFrame(sc_X.fit_transform(df.drop(["max", "tanggal", "categori", "critical", "location"], axis = 1)), 
                 columns=['pm10', 'pm25', 'so2', 'co', 'o3', 'no2'])
X.head()
```

```python
#create design matrix target vector y

import numpy as np

y = np.array(df.loc[:, ['categori']])

#split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
print('X_train {}, X_test {}, y_train {}, y_test {}'.format(len(X_train), len(X_test), len(y_train), len(y_test)))
```

```python
#Parameter tuning

#creating odd list of K for KNN
#Membuat List nilai ganjil K untuk KNN
myList = list(range(50))
neighbors = list(filter(lambda x: x%2 != 0, myList))

#empty list that will hold cv scores
#Membuat List kosong untuk menampung cv sores
cv_scores = []

#perform 10-fold cross validation
#Melakukan validasi silang 10 kali lipat
for k in neighbors:
  knn = KNeighborsClassifier(n_neighbors=k)
  scores = cross_val_score(knn, X_train, y_train.ravel(), cv=10, scoring='accuracy')
  cv_scores.append(scores.mean())

#Changing to missclassification error
MSE = [1-x for x in cv_scores]

#determining best k
optimal_k = neighbors[MSE.index(min(MSE))]
print('\nThe optimal number of neighbors is %d.' % optimal_k )
```

```python
#plot missclassification error vs k
plt.plot(neighbors, MSE)
plt.xlabel('Number of Neighbors K' )
plt.ylabel( 'MisclassificationError')
plt.show()
```

```python
#instantiate learning model (k=43)
knn = KNeighborsClassifier(n_neighbors=43)

#fitting the model
knn.fit(X_train, y_train.ravel())

#predict the response
pred = knn.predict(X_test)

pred
```

```python
#Menampilkan confusion mantrix hasil prediksi klasifikasi
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, pred))
```

```python
#Menampilkan visualisasi confusion matrix menggunakan Heatmap
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, pred)
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="Blues" ,fmt='g')
plt.title('Confusion matrix (k=9)')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
#Menampilkan hasil ketepatan prediksi klasifikasi

print(classification_report(y_test, pred))
```

```python
#evaluate accuracy
acc = accuracy_score(y_test, pred)*100
print('\nThe accuracy of the knn classifier for k=43 is %d%%' % acc)
```
